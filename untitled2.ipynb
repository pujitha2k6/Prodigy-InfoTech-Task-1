**STEP-1 IMPORT REQUIRED LIBRARIES**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**step-2 LOAD THE DATA SET**

```python
df = pd.read_csv("large_age_gender_population_dataset.csv")
```

**STEP-3 CREATE BAR CHART FOR GENDER DISTRIBUTION**

```python
# Group by gender
gender_distribution = df.groupby('Gender')['Population'].sum().reset_index()

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(data=gender_distribution, x='Gender', y='Population', palette='Set2')
plt.title('Total Population by Gender')
plt.xlabel('Gender')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**STEP-4 CREATE A BAR CHART FOR AGE GROUP DISTRIBUTION**

```python
# Group by age group
age_distribution = df.groupby('Age Group')['Population'].sum().reset_index()

# Optional: Sort the age groups logically
age_order = ['0-14', '15-24', '25-54', '55-64', '65+']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=age_distribution, x='Age Group', y='Population', order=age_order, palette='Blues')
plt.title('Total Population by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**STEP-5 GROUP DATA BY AGE GROUP AND GENDER**

```python
# Group by Age Group and Gender
grouped_data = df.groupby(['Age Group', 'Gender'])['Population'].sum().reset_index()

# Sort age groups for consistent order
age_order = ['0-14', '15-24', '25-54', '55-64', '65+']
grouped_data['Age Group'] = pd.Categorical(grouped_data['Age Group'], categories=age_order, ordered=True)
grouped_data = grouped_data.sort_values('Age Group')
```

**STEP-6 PLOT GROUPED BAR CHART**

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='Age Group', y='Population', hue='Gender', palette='Set1')

plt.title('Population Distribution by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.legend(title='Gender')
plt.show()
```

**STEP-7 ONE CHART PER COUNTRY**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("large_age_gender_population_dataset.csv")

# Ensure age group order
age_order = ['0-14', '15-24', '25-54', '55-64', '65+']
df['Age Group'] = pd.Categorical(df['Age Group'], categories=age_order, ordered=True)

# List of countries to visualize
selected_countries = ['India', 'China', 'United States', 'Brazil', 'Nigeria']  # you can change this list

# Loop through each country and plot
for country in selected_countries:
    country_data = df[df['Country'] == country]
    grouped = country_data.groupby(['Age Group', 'Gender'])['Population'].sum().reset_index()

    plt.figure(figsize=(8, 5))
    sns.barplot(data=grouped, x='Age Group', y='Population', hue='Gender', palette='Set2')
    plt.title(f'Population by Age Group and Gender in {country}')
    plt.xlabel('Age Group')
    plt.ylabel('Population')
    plt.legend(title='Gender')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()
```

**STEP-8 COMBINED CHART FOR FEW COUNTRIES**

```python
# Filter for selected countries
few_countries = ['India', 'China', 'United States']
filtered_df = df[df['Country'].isin(few_countries)]

# Group by Country, Age Group, and Gender
grouped = filtered_df.groupby(['Country', 'Age Group', 'Gender'])['Population'].sum().reset_index()
grouped['Age Group'] = pd.Categorical(grouped['Age Group'], categories=age_order, ordered=True)

# Plot
plt.figure(figsize=(12, 6))
sns.catplot(
    data=grouped,
    x='Age Group', y='Population', hue='Gender', col='Country',
    kind='bar', col_wrap=3, height=4, aspect=1.2,
    palette='Set1', sharey=False
)
plt.subplots_adjust(top=0.9)
plt.suptitle('Population Distribution by Age and Gender (Selected Countries)', fontsize=16)
plt.show()
```
